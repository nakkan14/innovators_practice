```python
import streamlit as st
from streamlit_folium import st_folium
import folium
import pandas as pd
import googlemaps
```

```python
# „Éá„Éº„Çø„Éô„Éº„Çπ„Å´Á∑ØÂ∫¶ÁµåÂ∫¶ÊÉÖÂ†±„ÇíËøΩÂä†
df = pd.read_csv('„ÅÑ„Åì„Éº„ÇàÂÖ¨Âúí.csv') 

gm = googlemaps.Client(key='APIKEY')

for i, r in df.iterrows():
    res = gm.geocode(r['address'])
    df.loc[i,'Á∑ØÂ∫¶'] = res[0]['geometry']['location']['lat']
    df.loc[i,'ÁµåÂ∫¶'] = res[0]['geometry']['location']['lng']
    print(i)
    print(res[0]['geometry']['location'])

df.to_csv('„ÅÑ„Åì„Éº„ÇàÂÖ¨Âúí.csv', index=None)
```

```python
# „Éö„Éº„Ç∏Ë®≠ÂÆö
st.set_page_config(
    page_title = "map",
    page_icon = "üóæ",
    layout = "wide"
)

# Âú∞Âõ≥„ÅÆ‰∏≠ÂøÉ„ÅÆÁ∑ØÂ∫¶/ÁµåÂ∫¶„ÄÅ„Çø„Ç§„É´„ÄÅÂàùÊúü„ÅÆ„Ç∫„Éº„É†„Çµ„Ç§„Ç∫„ÇíÊåáÂÆö„Åó„Åæ„Åô„ÄÇ
m = folium.Map(
    # Âú∞Âõ≥„ÅÆ‰∏≠ÂøÉ‰ΩçÁΩÆ„ÇíÂíåÊ≠åÂ±±Â∏Ç„Å´ÊåáÂÆö
    location = [34.23058133009892, 135.1719684167173], 
    # „Çø„Ç§„É´„ÄÅ„Ç¢„Éà„É™„Éì„É•„Éº„Éà„ÅÆÊåáÂÆö
    tiles = 'https://cyberjapandata.gsi.go.jp/xyz/pale/{z}/{x}/{y}.png',
    attr = 'Title„ÄÅFacility',
    # „Ç∫„Éº„É†„ÇíÊåáÂÆö
    zoom_start = 6
)

# Ë°®Á§∫„Åô„Çã„Éá„Éº„Çø„ÇíË™≠„ÅøËæº„Åø
df = pd.read_csv('„ÅÑ„Åì„Éº„ÇàÂÖ¨Âúí.csv')

print(df.columns)
```

```python

# Ë™≠„ÅøËæº„Çì„Å†„Éá„Éº„Çø(Á∑ØÂ∫¶„ÉªÁµåÂ∫¶„ÄÅ„Éù„ÉÉ„Éó„Ç¢„ÉÉ„ÉóÁî®ÊñáÂ≠ó„ÄÅ„Ç¢„Ç§„Ç≥„É≥„ÇíË°®Á§∫)
for i, row in df.iterrows():
    # „Éù„ÉÉ„Éó„Ç¢„ÉÉ„Éó„ÅÆ‰ΩúÊàê(ÈÉΩÈÅìÂ∫úÁúåÂêçÔºãÈÉΩÈÅìÂ∫úÁúåÂ∫ÅÊâÄÂú®Âú∞Ôºã‰∫∫Âè£ÔºãÈù¢Á©ç)
    pop = f"{row['title']}<br>{row['facility']}"
    folium.Marker(
        # Á∑ØÂ∫¶„Å®ÁµåÂ∫¶„ÇíÊåáÂÆö
        location=[row['Á∑ØÂ∫¶'], row['ÁµåÂ∫¶']],
        # „ÉÑ„Éº„É´„ÉÅ„ÉÉ„Éó„ÅÆÊåáÂÆö(ÈÉΩÈÅìÂ∫úÁúåÂêç)
        tooltip=row['title'],
        # „Éù„ÉÉ„Éó„Ç¢„ÉÉ„Éó„ÅÆÊåáÂÆö
        popup=folium.Popup(pop, max_width=300),
        # „Ç¢„Ç§„Ç≥„É≥„ÅÆÊåáÂÆö(„Ç¢„Ç§„Ç≥„É≥„ÄÅËâ≤)
        icon=folium.Icon(icon="tree-conifer",icon_color="white", color="green")
    ).add_to(m)

st_data = st_folium(m, width=1200, height=800)

```
